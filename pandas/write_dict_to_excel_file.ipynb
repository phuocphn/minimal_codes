```python
# Construct DataFrame from a list of dictionaries

data = [{'Stimulus Time': 1, 'Reaction Time': 1},
        {'Stimulus Time': 2, 'Reaction Time': 2},
        {'Stimulus Time': 3, 'Reaction Time': 3},
        {'Stimulus Time': 4, 'Reaction Time': 4}]
df = DataFrame(data)
df
```

```python
df.to_csv('test_from_dict.csv', index=False)
```
